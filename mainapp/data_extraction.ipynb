```python
def format_number(number):
    number = int(number)
    if number >= 10000000:  # If number is greater than or equal to 10 crores
        return "{:,.2f} crores".format(number / 10000000)
    elif number >= 100000:  # If number is greater than or equal to 1 lakh
        return "{:,.2f} lakhs".format(number / 100000)
    elif number >= 1000:  # If number is greater than or equal to 1 thousand
        return "{:,.2f} thousands".format(number / 1000)
    else:
        return "{:,.2f}".format(number)
```

```python
import json
import pandas as pd
aggreated_transaction_path = 'D:/Learning/Projects/PhonepePulseDataVisualization/src/phonepeData/data/aggregated/transaction/country/india/'

transaction ={'Year':[],'Quarter':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}

year_list = os.listdir(aggreated_transaction_path)

for year in year_list:
    if year  == 'state':
        continue
    path = aggreated_transaction_path+year+'/'
    quarter_list = os.listdir(path)
    for quarter in quarter_list:
        qtr_path = path+quarter
        D = open(qtr_path,'r')
        data =json.load(D)
        for value in  data['data']['transactionData']:
            Name = value['name']
            count = value['paymentInstruments'][0]['count']
            amount=format_number(value['paymentInstruments'][0]['amount'])
            transaction['Transaction_amount'].append(amount)
            transaction['Transaction_count'].append(count)
            transaction['Transaction_type'].append(Name)
            transaction['Year'].append(year)
            transaction['Quarter'].append(int(quarter.strip('.json')))

transcation_df = pd.DataFrame(transaction)         

    
```
